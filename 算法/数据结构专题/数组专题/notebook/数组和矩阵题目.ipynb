æ¯”è¾ƒå®¹æ˜“æƒ³åˆ°çš„ä¸¤ç§æ€è·¯ï¼š1ã€å®šä¹‰curRowå’ŒcurColè®°å½•å½“å‰éå†åˆ°çš„è¡Œå’Œåˆ—ï¼›éœ€è¦è€ƒè™‘åªå‰©ä¸€è¡Œå’Œåªå‰©ä¸€åˆ—çš„æƒ…å†µï¼›2ã€å®šä¹‰tRã€tCã€dRå’ŒdCå®šä½å½“å‰éå†åˆ°çš„å­çŸ©é˜µï¼Œä¸€å±‚ä¸€å±‚å¾€é‡Œç¼©å°èŒƒå›´ï¼›ä¹Ÿéœ€è¦å•ç‹¬è€ƒè™‘åªå‰©ä¸€è¡Œå’Œåªå‰©ä¸€åˆ—çš„æƒ…å†µã€‚

```python
def spiralOrderPrint(matrix):
    def traceMatrix(tR, tC, dR, dC):
        # å¦‚æœçŸ©é˜µåªæœ‰ä¸€è¡Œ
        if tR == dR:
            for i in range(tC, dC + 1):
                ans.append(matrix[tR][i])
        # å¦‚æœçŸ©é˜µåªæœ‰ä¸€åˆ—
        elif tC == dC:
            for i in range(tR, dR + 1):
                ans.append(matrix[i][tC])
        else:
            curR, curC = tR, tC
            while curC != dC:
                ans.append(matrix[tR][curC])
                curC += 1
            while curR != dR:
                ans.append(matrix[curR][dC])
                curR += 1
            while curC != tC:
                ans.append(matrix[dR][curC])
                curC -= 1
            while curR != tR:
                ans.append(matrix[curR][tC])
                curR -= 1

    ans = []
    tR, tC, dR, dC = 0, 0, len(matrix) - 1, len(matrix[0]) - 1
    while tR <= dR and tC <= dC:
        traceMatrix(tR, tC, dR, dC)
        tR, tC, dR, dC = tR + 1, tC + 1, dR - 1, dC - 1
    return ans
```

é¢˜ç›®ï¼šç»™å®šä¸€ä¸ªnxnçš„matrixï¼ŒæŠŠè¿™ä¸ªçŸ©é˜µåŸåœ°æ—‹è½¬90åº¦ï¼Œè¦æ±‚ç©ºé—´å¤æ‚åº¦O(1).

```python
def rotate(matrix):
    def rotate_submatrix(tR, tC, dR, dC):
        times = dC - tC
        for i in range(times):
            temp = matrix[tR][tC + i]
            matrix[tR][tC + i] = matrix[dR - i][tC]
            matrix[dR - i][tC] = matrix[dR][dC - i]
            matrix[dR][dC - i] = matrix[tR + i][dC]
            matrix[tR + i][dC] = temp
        
        
    tR, tC, dR, dC = 0, 0, len(matrix) - 1, len(matrix[0]) - 1
    while tR < dR:
        rotate_submatrix(tR, tC, dR, dC)
        tR, tC, dR, dC = tR + 1, tC + 1, dR - 1, dC - 1
```

ä¸‹é¢è¿™ä¸ªè§£æ³•æ¯”è¾ƒç¬¨ï¼Œä½†èƒ½å®ç°åŠŸèƒ½

```python
def printMatrixZigZag(matrix):
    ans = []
    curR, curC = 0, 0
    direction = 0
    while curR != len(matrix) - 1 or curC != len(matrix[0]) - 1:
        # å‘å³ä¸Šæ–¹è¿åŠ¨
        if direction == 0:
            while curR >= 0 and curC < len(matrix[0]):
                ans.append(matrix[curR][curC])
                curR -= 1
                curC += 1
            direction = 1
            if curC < len(matrix[0]):
                curR = 0
            else:
                curR += 2
                curC = len(matrix[0]) - 1
        # å‘å·¦ä¸‹æ–¹è¿åŠ¨
        else:
            while curR < len(matrix) and curC >= 0:
                ans.append(matrix[curR][curC])
                curR += 1
                curC -= 1
            direction = 0
            if curR < len(matrix) :
                curC = 0
            else:
                curC += 2
                curR = len(matrix) - 1
    ans.append(matrix[len(matrix) - 1][len(matrix[0]) - 1])
    return ans
```

```python
import numpy as np

testArr = np.arange(10).reshape(2,5) + 1
print(testArr)
printMatrixZigZag(testArr)
```
